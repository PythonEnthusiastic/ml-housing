```python
# To nagvigate the filesystem
import os
# To unzip the tar file
import tarfile
# To download data
import urllib
```

```python
# URL of the data we will be installing, keep in mind it is a tarfile
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
# State the file path of where the data will live (get it it? live)
HOUSING_PATH = os.path.join("datasets", "housing")
# The URL of the actual data
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
```

```python
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    # Create new file
    os.makedirs(housing_path, exist_ok=True)
    # State the path of the installed data
    tgz_path = os.path.join(housing_path, 'housing.tgz')
    # Download the data from the url into the path provided
    urllib.request.urlretrieve(housing_url, tgz_path)
    # Open the tarfile
    housing_tgz = tarfile.open(tgz_path)
    # Extract all its content
    housing_tgz.extractall(path=housing_path)
    # Close the tarfile
    housing_tgz.close()
```

```python
fetch_housing_data()
```
